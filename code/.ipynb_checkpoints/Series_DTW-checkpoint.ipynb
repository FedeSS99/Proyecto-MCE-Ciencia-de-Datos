Carga de librerías

```python
import matplotlib.pyplot as plt
import numpy as np

from collections import Counter

from Rutinas_DTW import DTW, MatrixDTW 
```

Lectura de archivos con datos de entrenamiento y prueba

```python
with open("../StarLightCurves/StarLightCurves_TRAIN.txt", "r") as input_data:
    class_series_train = dict()
    for n, line in enumerate(input_data.readlines()):
        line_split = line.split("  ")[1:]
        class_series, data_series = int(line_split[0][1]), np.array([float(val_string) for val_string in line_split[1:]])
        class_series_train[n] = (class_series, data_series)

with open("../StarLightCurves/StarLightCurves_TEST.txt", "r") as input_data:
    class_series_test = dict()
    for n, line in enumerate(input_data.readlines()):
        line_split = line.split("  ")[1:]
        class_series, data_series = int(line_split[0][1]), np.array([float(val_string) for val_string in line_split[1:]])
        class_series_test[n] = (class_series, data_series)  
```

Obtención de clases por serie de tiempo para el par de conjuntos de entrenamiento y prueba

```python
classes_train, _ = zip(*class_series_train.values())
classes_train_counter = dict(Counter(classes_train))

classes_test, _ = zip(*class_series_test.values())
classes_test_counter = dict(Counter(classes_test))

print(classes_train_counter, sum(classes_train_counter.values()))
print(classes_test_counter, sum(classes_test_counter.values()))
```

## Visualización de series de tiempo

Asignación de color para cada clase

```python
colors = {1: "red", 2: "green", 3: "blue"}
```

```python

Figure, Subplot = plt.subplots(1, 3, sharey="all", figsize=(12,6))
Figure.suptitle("Train dataset")

for class_data, series_data in class_series_train.values():
    Subplot[class_data - 1].plot(series_data, alpha = 0.1, color = colors[class_data])

for k in range(1,4):
    Subplot[k-1].set_title(f"Class {k}")
    Subplot[k-1].set_xlabel("Time index")
    Subplot[k-1].set_ylabel("Amplitude")

plt.show()
```

```python
Figure, Subplot = plt.subplots(1, 3, sharey="all", figsize=(12,6))
Figure.suptitle("Test dataset")

for class_data, series_data in class_series_test.values():
    Subplot[class_data - 1].plot(series_data, alpha = 0.05, color = colors[class_data])

for k in range(1,4):
    Subplot[k-1].set_title(f"Class {k}")
    Subplot[k-1].set_xlabel("Time index")
    Subplot[k-1].set_ylabel("Amplitude")
```

## Dynamic Time Warping

```python
_, train_series = zip(*class_series_train.values())
_, test_series = zip(*class_series_test.values())
```

```python
%%timeit

DTW(train_series[0], train_series[1], q=2.0)
```

```python
input_series = np.array(list(test_series))

dtw_series = MatrixDTW(input_series, 2.0)
```

```python
with open("../output data/DWT.npy", "+w") as output_dtw:
    np.save(output_dtw, dtw_series)
```

```python
Figure_DTWmatrix = plt.figure(figsize=(8, 8))
Subplot_DTW = Figure_DTWmatrix.add_subplot(111)

DTW_image = Subplot_DTW.imshow(dtw_series, 
                               cmap = "nipy_spectral", 
                               aspect = "equal", 
                               origin = "lower")
Subplot_DTW.set_ylabel("n-th index series")
Subplot_DTW.set_xlabel("m-th index series")
plt.colorbar(DTW_image, ax = Subplot_DTW, label = "DTW score")

plt.show()
```
